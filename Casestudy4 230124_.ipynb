```python
ct_df1['txn_date'] = pd.to_datetime(ct_df1['txn_date'])
```

```python
cn_df1['start_date'] = pd.to_datetime(cn_df1['start_date'])

cn_df1['end_date'] = pd.to_datetime(cn_df1['end_date'], errors='coerce')
```

```python
merged_df = pd.merge(r_df1, cn_df1, on='region_id', how='inner')

final_merged_df = pd.merge(merged_df, ct_df1, on='customer_id', how='inner')

final_merged_df
```

```python
# How many unique nodes are there on the Data Bank system?

unique_nodes = cn_df1['node_id'].nunique()

unique_nodes
```

```python
# What is the number of nodes per region?
nodes_per_region = cn_df1.groupby('region_id')['node_id'].nunique().reset_index()

print("\nNodes per Region:")

nodes_per_region
```

```python
# How many customers are allocated to each region?

customers_per_region = cn_df1.groupby('region_id')['customer_id'].nunique().reset_index()

print("\nCustomers per Region:")

customers_per_region
```

```python
# How many days on average are customers reallocated to a different node?

avg_reallocation_days = (cn_df1['end_date'] - cn_df1['start_date']).mean().days

print(f"\nAverage Reallocation Days: {avg_reallocation_days} days")
```

```python
# What is the median, 80th and 95th percentile for this same reallocation days metric for each region?

cn_df1['reallocation_days'] = (cn_df1['end_date'] - cn_df1['start_date']).dt.days

reallocation_days_percentiles = cn_df1.groupby('region_id')['reallocation_days'].quantile([0.5, 0.8, 0.95]).reset_index()

reallocation_days_percentiles
```

```python
# What is the unique count and total amount for each transaction type?
transaction_stats = ct_df1.groupby('txn_type').agg({'customer_id': 'nunique', 'txn_amount': 'sum'}).reset_index()
transaction_stats.columns = ['txn_type', 'unique_count', 'total_amount']
print("\nTransaction Statistics:")
transaction_stats
```

```python
# What is the average total historical deposit counts and amounts for all customers?

avg_deposit_counts = ct_df1[ct_df1['txn_type'] == 'deposit'].groupby('customer_id')['txn_date'].count().mean()
avg_deposit_amount = ct_df1[ct_df1['txn_type'] == 'deposit'].groupby('customer_id')['txn_amount'].sum().mean()

print(f"\nAverage Deposit Counts: {avg_deposit_counts}")
print(f"Average Deposit Amount: {avg_deposit_amount}")
```

```python
# For each month - how many Data Bank customers make more than 1 deposit and either 1 purchase or 1 withdrawal in a single month?

monthly_activity = (ct_df1[(ct_df1['txn_type'] == 'deposit') &
                           (ct_df1['customer_id'].isin(ct_df1[ct_df1['txn_type']
                                                              .isin(['purchase', 'withdrawal'])].customer_id.unique()))]
                    .groupby(ct_df1['txn_date'].dt.to_period("M"))
                    .agg({'customer_id': 'nunique'}).reset_index())
monthly_activity.columns = ['month', 'customers_count']
print("\nMonthly Activity:")
monthly_activity
```

```python
# What is the closing balance for each customer at the end of the month?

ct_df1['month'] = ct_df1['txn_date'].dt.to_period("M")

closing_balance = (ct_df1.groupby(['customer_id', 'month'])
                   .apply(lambda x: x['txn_amount'] *
                          (-1 if x['txn_type'].isin(['purchase', 'withdrawal']).any() else 1))
                   .groupby(['customer_id', 'month'])
                   .sum().reset_index())

closing_balance.columns = ['customer_id', 'month', 'closing_balance']

print("\nClosing Balance:")

closing_balance
```

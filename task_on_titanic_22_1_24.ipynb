### Identifing and handling outliers in fare with appropriate techniques


```python
sns.boxplot(df['fare'])
```

```python
def luFence(x):# function for inter quartile range
    #x = x.sort()
    iqr = x.quantile(0.75) - x.quantile(0.25)
    lowerFence, upperFence = max(min(x), x.quantile(0.25)-1.5*iqr), min(max(x), x.quantile(0.75)+1.5*iqr)
    return lowerFence, upperFence
```

```python
luFence(df['fare'])
```

```python
outliers = (df['fare'] < 0.0) | (df['fare'] > 65.6344)
```

### imputing missing values with mode in 'Fare' column


```python
fare_mode = df['fare'].mode()
df.loc[outliers, 'fare'] = fare_mode

print("Number of outliers replaced:", outliers.sum())
```

```python
df['clipped_data_of_fare'] = df['fare'].clip(lower = 0.0, upper = 65.6344)
sns.boxplot(x='clipped_data_of_fare', data=df)
```

### imputing missing values with median in 'Fare' column

```python
fare_median = df['fare'].median()
df.loc[outliers, 'fare'] = fare_median

print("Number of outliers replaced:", outliers.sum())
```

### imputing missing values with mode in 'Fare' columnÂ¶

```python
fare_mean = df['fare'].mean()
df.loc[outliers, 'fare'] = fare_mean

print("Number of outliers replaced:", outliers.sum())
```

* finally concluded that Imputing Fare values is the appropriate method for cleaing Outliers
